```python
def pcaCompress(img, n_components, trancat_col=True, channels=True):

    if not trancat_col:
        img = np.einsum('jik->ijk', img)
    print('img_shape ',img.shape)
    
    output = []
    for i in range(3):
        img_ = img[:,:,i]
        pca_ = PCA(n_components=n_components)
        img_ = pca_.fit_transform(img_)
        img_ = pca_.inverse_transform(img_)
        output.append(img_)

    return np.stack(output, axis=2)
```

```python
img_compressed = pcaCompress(img, 64)
```

```python
# plt.imshow(cv2.cvtColor(np.array(img_compressed, dtype=int), cv2.COLOR_BGR2RGB))
cv2.imwrite('./quantized_k_10_image5_compressed.jpg', img_compressed)
```
