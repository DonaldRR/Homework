```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
from tqdm import tqdm
import pandas as pd
```

```python
f = open('balance-scale.data.txt')
data = []
labels = []
```

```python
for line in tqdm(f.readlines()):
    line = line[:-1].split(',')
    data.append(line[1:])
    labels.append(line[0])
```

```python
data = np.array(data, dtype=int)
```

```python
data
```

```python
labels = np.reshape(np.array(labels), (len(labels),1))
```

```python
data.shape
```

```python
labels.shape
```

```python
df = pd.DataFrame(data, columns=['Left-Weight','Left-Distance','Right-Weight','Right-Distance'])
```

```python
df['labels'] = labels
```

```python
df
```

```python
df.to_csv('balance_scale.csv')
```
